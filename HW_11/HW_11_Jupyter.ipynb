```python
from sympy.plotting import plot
from sympy import solve
import sympy
from sympy import Symbol
```

```python
x = Symbol('x')
f = x**3 - 50*x
f
```

```python
g = -x**4 + 88*x**2 - 241
g
```

```python
h = f - g
plot(f,g,h)
```

```python
k = solve(h)
[[x_elem.n().round(2),f.subs(x,x_elem).n().round(2)] for x_elem in k]
```
